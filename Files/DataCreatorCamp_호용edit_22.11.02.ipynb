```python
## 수행시간 약 11분 ##
root_dir = '/content/drive/MyDrive/Colab Notebooks/DataCreatorCamp/DataSet'
folder_list = os.listdir(root_dir) #데이터셋 속의 폴더가 여러개이므로 리스트로 만듦
folder_list_len = len(folder_list) #데이터셋 속의 폴더 갯수
dataset = []
inputscale=0 #총 데이터 크기 저장
cnt =1
for i in range(folder_list_len):
  tmpdatadir1 = root_dir + '/' + folder_list[i] + '/' # 각 폴더 경로 저장
  inputscale += len(os.listdir(tmpdatadir1))
print(inputscale)

for j in range(folder_list_len):

  tmpdatadir = root_dir + '/' + folder_list[j] + '/' # 각 폴더 경로 저장
  data_dir = tmpdatadir #데이터 파일의 경로 -> 방금 만든 폴더의 경로
  data_list = os.listdir(data_dir) #각 폴더 속 모든 파일 불러오기
  #print(tmpdatadir)
  data_list_jpg = [img for img in data_list if img.endswith(".jpg")] #jpg확장자만 불러오기
  #print(data_list_jpg)
  for k in range(len(data_list_jpg)):
    datatmp = data_list_jpg[k]
    image = cv2.imread(datatmp, cv2.IMREAD_COLOR) #파일 하나하나 읽어오기 
    #reimage = cv2.resize(image,(100,100))
    # 확장자 추가로 확인하기
    data = np.array(image) #파일을 넘파이 배열로 변환
    dataset.append(data) #변환한 넘파이 배열을 데이터셋에 추가
    print(cnt,"번째 데이터",datatmp,"추가 완료")
    cnt+=1
np.save('/content/drive/MyDrive/Colab Notebooks/DataCreatorCamp/DataSet.npy',dataset)
```
